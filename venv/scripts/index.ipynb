# Import Dataset
we import data from kaggle
data is an iris dataset
iris is a species of flower

```python
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv("../datasets/iris.csv")
df
```

```python
df2 = pd.DataFrame(df)
df2
```

```python
df3 = df2
df3.columns = ['SepalLength', 'SepalWidth', 'PetalLength','PetalWidth','Species']
df3
```

```python
df4 = df3
df4.isnull().sum()
```

```python
sns.countplot(x='Species', data = df4)#calculates the amount of data per class, as the x axis
```

```python
fig1 = sns.scatterplot(x = 'SepalLength',y = 'SepalWidth',hue = 'Species', data = df4)
fig1.set(title = 'Sepal Length & Sepal Width')
```

```python
import plotly.express as px
px.histogram(df4, x='Species',color='Species')
```

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn = KNeighborsClassifier(n_neighbors= 5, weights="uniform") 
x_train = df4.drop(['Species'], axis=1)
x_train
```

```python
y_train = df4.drop(['SepalLength','SepalWidth','PetalLength','PetalWidth'],axis=1)
y_train
```

```python
type(y_train)
# convert ytrain from dataframe to 1darray
```

```python
knn.fit(x_train, y_train)
train_acc = knn.score(x_train,y_train)
```
